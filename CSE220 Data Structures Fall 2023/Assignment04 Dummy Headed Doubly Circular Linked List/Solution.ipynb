```python
class Patient:
    
    def __init__(self,id,name,age,bloodgroup):
        
        self.id=id
        self.name=name
        self.age=age
        self.blood_group=bloodgroup
        self.prev=None
        self.next=None



class WRM:
    
    def __init__(self):
        
        self.head=Patient(None,None,None,None)
        self.head.prev=self.head
        self.head.next=self.head
        self.tail=self.head

    def RegisterPatient(self,id,name,age,bloodgroup):
        
        if self.head.next==self.head:
            self.tail=self.head

        new_patient=Patient(id,name,age,bloodgroup)
        new_patient.next=self.head
        new_patient.prev=self.tail
        self.tail.next=new_patient
        self.tail=new_patient
        self.head.prev=new_patient

    def ServePatient(self):
        
        if self.head.next==self.head:
            print(f"No patients in the waiting room.")
        else:
            temp=self.head.next
            print(f'{temp.name} is served' )
            self.head.next=temp.next
            temp.next.prev=self.head
            return self.head
        
    def CancelAll(self):
        
        if self.head.next==self.head:
            print("No aapointment to be cancelled")
        else:
            temp=self.head.next
            while temp.next != self.head:
                self.head.next = temp.next
                temp.next.prev = self.head
                temp = temp.next
            self.head.next = self.head
            self.head.prev = self.head
            return self.head
        
    def CanDoctorGoHome(self):
        
        if self.head.next == self.head:
            return True
        else:
            return False

    def ShowAllPatient(self):
        
        if self.head.next == self.head:
            print(f"No patient in the waiting room.")
        else:
            temp = self.head.next
            while temp != self.head:
                print(temp.name, end = " ")
                temp = temp.next
            print()

    def ReverseTheLine(self):
        
        temp = self.head.next
        while temp.name != None:
            save1 = temp.next
            save2 = temp.prev
            temp.next = save2
            temp.prev = save1
            temp = save1
        save = self.head.next
        self.head.next = self.head.prev
        self.head.prev= save
        return self.head



wrm = WRM()
while True:
  print(f"**Welcome to Waiting Room Management System**")
  print(f"==Choose an option")
  print(f"1. Add Patient")
  print(f"2. Serve Patient")
  print(f"3. Show all patients")
  print(f"4. Can doctor go home?")
  print(f"5. Cancel all appointments")
  print(f"6. Reverse the line")
  print(f"7. Exit")
  x = int(input("Enter your choice: "))
  if x == 1:
    id = input("Enter ID: ")
    name = input("Enter Name: ")
    age = input("Enter Age: ")
    bg = input("Enter Blood Group: ")
    wrm.RegisterPatient(id, name, age, bg)
    print(f"Patient registered successfully!")
    print()
  elif x == 2:
    wrm.ServePatient()
    print()
  elif x == 3:
    wrm.ShowAllPatient()
    print()
  elif x == 4:
    if wrm.CanDoctorGoHome()== True:
      print(f"Yes, the doctor can go home.")
    else:
      print(f"No, the doctor can't go home.")
    print()
  elif x == 5:
    wrm.CancelAll()
    print(f"All appointments cancelled.")
    print()
  elif x == 6:
    wrm.ReverseTheLine()
    print(f"Successfully reversed!")
  elif x == 7:
    print(f"Thank you for using Waiting Room Management.")
    break
  else:
    print(f"No such option")
    print()
```
